```python
import pandas as pd
```

```python
file_path = "AnnualAverageRainFall.csv"
df = pd.read_csv(file_path)
df
```

```python
df.info()
```

```python
df.rename(columns={df.columns[0]: 'District'}, inplace=True)
df
```

```python
df.dropna(subset=df.columns[1:], how='all', inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df
```

```python
df.rename(columns={df.columns[0]: "District"}, inplace=True)
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
df.fillna(df.mean(numeric_only=True), inplace=True)
df.dropna(thresh=len(df.columns) * 0.7, inplace=True)
df.drop_duplicates(subset=["District"], keep="first", inplace=True)
df.to_csv("Cleaned_AnnualAverageRainFall.csv", index=False)
print("Data cleaning completed. Cleaned file saved.")
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

# Select a district for visualization
district = "Gurdaspur"

# Plot rainfall over years for the district
plt.figure(figsize=(12, 6))
plt.plot(df.columns[1:], df[df['District'] == district].values.flatten()[1:],
         marker='o', linestyle='-', label=district, color='b')

plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.title(f"Annual Rainfall Trend for {district}")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the average rainfall for each district
df_avg = df.set_index("District").mean(axis=1).sort_values()

# Bar chart
plt.figure(figsize=(12, 6))
df_avg.plot(kind='bar', color='skyblue')

plt.xlabel("District")
plt.ylabel("Average Rainfall (mm)")
plt.title("Average Rainfall Across Districts")
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Reload the original dataset to show missing values
df_original = pd.read_csv("AnnualAverageRainFall.csv")

# Heatmap of missing values
plt.figure(figsize=(10, 5))
sns.heatmap(df_original.isnull(), cmap="viridis", cbar=False)

plt.title("Missing Values in Original Dataset")
plt.show()
```

```python
import seaborn as sns

# Convert data to long format for Seaborn
df_melted = df.melt(id_vars=["District"], var_name="Year", value_name="Rainfall")

# Box plot
plt.figure(figsize=(14, 6))
sns.boxplot(x="Year", y="Rainfall", data=df_melted)

plt.xticks(rotation=90)
plt.title("Rainfall Distribution Across Years")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

X, Y = np.meshgrid(range(df.shape[1]-1), range(df.shape[0]))
Z = df.iloc[:, 1:].values

ax.plot_surface(X, Y, Z, cmap="coolwarm")
ax.set_xlabel("Years")
ax.set_ylabel("Districts")
ax.set_zlabel("Rainfall (mm)")
plt.title("3D Surface Plot of Rainfall Variation")
plt.show()
```

```python
top_n = 5  # Number of top districts to show

df_top = df_avg.nlargest(top_n)
df_other = pd.Series([df_avg.iloc[top_n:].sum()], index=["Others"])

df_combined = pd.concat([df_top, df_other])

plt.figure(figsize=(8, 6))
df_combined.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='viridis')
plt.ylabel("")
plt.title("Top 5 Districts + Others in Rainfall Contribution")
plt.show()
```

```python
year1 = "2000"
year2 = "2020"

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

df.set_index("District")[year1].nlargest(10).plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='coolwarm', ax=axes[0])
axes[0].set_ylabel("")
axes[0].set_title(f"Top 10 Districts in {year1}")

df.set_index("District")[year2].nlargest(10).plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='coolwarm', ax=axes[1])
axes[1].set_ylabel("")
axes[1].set_title(f"Top 10 Districts in {year2}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_avg.nlargest(10).plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='coolwarm', wedgeprops={'edgecolor': 'black'})

# Add a circle at the center to create a "donut" effect
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

plt.ylabel("")
plt.title("Donut Chart: Top 10 Districts with Highest Rainfall")
plt.show()
```
